```python
import matplotlib.pyplot as plt

# 
matrix_sizes = [128, 256, 512, 1024, 1536, 2048, 3072, 4096]

#  kernel  FLOP/second 
# 
cublas_flops = [0.6e12, 2.86e12, 6.45e12, 7.67e12, 8.13e12, 7.97e12, 8.29e12, 8.25e12]
vectorize_flops = [0.15e12, 0.74e12, 3.05e12, 5.77e12, 7.33e12, 7.30e12, 7.50e12, 6.53e12]
warp_tile_flops = [0.15e12, 0.68e12, 2.78e12, 5.23e12, 6.16e12, 6.21e12, 6.27e12, 5.47e12]
double_buffering_flops = [0.16e12, 0.73e12, 3.17e12, 4.44e12, 5.05e12, 5.21e12,5.04e12, 4.71e12]

# 
plt.figure(figsize=(10, 6))

plt.plot(matrix_sizes, cublas_flops, marker='o', linestyle='-', label='cublas', color='b')
plt.plot(matrix_sizes, vectorize_flops, marker='s', linestyle='--', label='vectorize', color='g')
plt.plot(matrix_sizes, warp_tile_flops, marker='^', linestyle='-.', label='warp tile', color='r')
plt.plot(matrix_sizes, double_buffering_flops, marker='p', linestyle=':', label='double_buffer', color='m')

# 
plt.title('Different Kernel Comparison', fontsize=16)
plt.xlabel('Matrix Size (side length)', fontsize=14)
plt.ylabel('FLOP/second', fontsize=14)

# 
plt.xticks(matrix_sizes, fontsize=9)
plt.yticks(fontsize=12)

# 
plt.grid(True, linestyle='--', alpha=0.6)

# 
plt.legend(fontsize=12)

# 
plt.tight_layout()
plt.show()
```
